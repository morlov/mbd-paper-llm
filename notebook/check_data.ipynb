```python
import pandas as pd

EMB_PATH = f'../data/embeddings/baselines/'
TARGETS_PATH = '../data/mbd_targets_balanced.parquet'

def format_embedding_data_and_save(emb_name):
    
    df_emb = pd.read_parquet(EMB_PATH + emb_name + '/')

    emb_i = 0

    renamed = dict()
    for i, col in enumerate(df_emb.columns):
        renamed[col] = col
        if col.startswith("emb_"):
            renamed[col]  = f'embed_{emb_i}'
            emb_i += 1

    df_emb.client_id = df_emb.client_id.apply(lambda x: x.split('_')[0])
    df_emb.rename(columns=renamed, inplace=True)
    df_emb.drop(['target_1', 'target_2', 'target_3', 'target_4'], axis=1, inplace=True)

    df_targets = pd.read_parquet(TARGETS_PATH, columns=['client_id'])

    df = df_emb.merge(df_targets, on="client_id")
    df.to_parquet(EMB_PATH + f'/{emb_name}.parquet', index=False)
    print(df.shape)
    
    return df
```
